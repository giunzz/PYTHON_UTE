```python
df = pd.read_csv('shopping_trends.csv', header=None) # header=None: tệp CSV không có hàng tiêu đề (header). Vì vậy, Pandas sẽ tự động gán các cột theo số thứ tự (0, 1, 2, ...).

# Rename columns : được sử dụng để thay đổi tên cột trong DataFrame.
df = df.rename(columns={
    0: 'Customer ID',
    1: 'Age',
    2: 'Gender',
    3: 'Item Purchased',
    4: 'Category',
    5: 'Purchase Amount (USD)',
    6: 'Location',
    7: 'Size',
    8: 'Color',
    9: 'Season',
    10: 'Review Rating',
    11: 'Subscription Status',
    12: 'Payment Method',
    13: 'Shipping Type',
    14: 'Discount Applied',
    15: 'Promo Code Used',
    16: 'Previous Purchases',
    17: 'Preferred Payment Method',
    18: 'Frequency of Purchases'
})
df.head()
```

```python
#delete row

df = df.drop(df[df['Customer ID'] == '3'].index) # xóa hàng có giá trị cột 'Customer ID' = 1
df.head()
```

```python
#add row
df.loc[len(df)] = ['1', '22', 'M', 'Shirt', 'Clothing', '50', 'New York', 'M', 'Blue', 'Summer', '5', 'Active', 'Credit Card', 'Standard', 'Yes', 'No', '3', 'Credit Card', 'Monthly']
# len(df): Lấy độ dài hiện tại của DataFrame, tức là số hàng hiện có trong df.
# df.loc[len(df)]: Thêm một hàng mới tại vị trí cuối cùng (vị trí tiếp theo sau hàng cuối cùng của DataFrame hiện tại).
```

**Import ploting library**

```python
import seaborn as sns # thư viện trực quan hóa dữ liệu (biểu đồ histogram, scatter plot, box plot, bar plot, heatmap,...)
import matplotlib.pyplot as plt # thư viện trực quan hóa dữ liệu (biểu đồ line, bar, scatter, histogram, boxplot, pie, v.v.)
# Matplotlib và Seaborn là 2 thư viên được sử dụng rộng rãi để trực quan hoá dữ liệu dạng bảng
```

**Distribution of Item Purchase**


```python
plt.figure(figsize=(20, 8)) # tạo một figure với kích thước 20x8
#Các biểu đồ trong matplotlib sẽ nằm trong một Figure object.
df['Item Purchased'].value_counts().plot(kind='pie',autopct='%1.1f%%')
# df['Item Purchased'].value_counts(): Đếm số lần xuất hiện của từng giá trị trong cột Item Purchased.
# kind='pie': Vẽ biểu đồ tròn.
# autopct='%1.1f%% Hiển thị phần trăm trên mỗi phần của biểu đồ. Hiển thị số thập phân với 1 chữ số.
plt.title('Distribution of Item purchased') # title của biểu đồ
plt.ylabel("") # Xóa nhãn trục Y
plt.tight_layout() # Điều chỉnh bố cục để tránh các phần bị cắt.
plt.show() # Hiển thị biểu đồ
# Biểu đồ hiện thị phần trăm số lần mua của từng loại hàng trong cột 'Item Purchased'
```

```python
item = df.groupby('Item Purchased')['Purchase Amount (USD)'].sum().reset_index()
# groupby('Item Purchased'): Nhóm dữ liệu theo giá trị cột 'Item Purchased'.
# ['Purchase Amount (USD)']: Chọn cột 'Purchase Amount (USD)'.
# sum(): Tổng số tiền đã chi cho từng mục Purchase Amount (USD).
# reset_index(): Đặt lại chỉ số của DataFrame.
# Hiển thị tổng số tiền đã chi cho từng mục Purchase Amount (USD) cho từng mục Item Purchased.
```

**Item purchased by Amount(USD)**

```python
plt.figure(figsize=(12, 5)) # tạo một figure với kích thước 12x5
sns.barplot(data=item,x='Item Purchased',y='Purchase Amount (USD)') # vẽ biểu đồ barplot với dữ liệu item, trục x là 'Item Purchased', trục y là 'Purchase Amount (USD)'
plt.title('Item purchased by Amount(USD)') # title của biểu đồ
plt.ylabel('Purchase Amount (USD)') # nhãn trục y
plt.xticks(rotation=90) # Xoay nhãn trục x 90 độ
plt.tight_layout() # Điều chỉnh bố cục để tránh các phần bị cắt.
plt.show() 
```
